```python
trig_index = np.arange(-5,5,.1)
```

```python
sin_arr = np.sin(trig_index)
cos_arr = np.cos(trig_index)
tan_arr = np.tan(trig_index)
```

```python
trig_df = pd.DataFrame({"sin(x)":sin_arr,"cos(x)":cos_arr, "tan(x)":tan_arr},index=trig_index)
trig_df.index.name = "x"
trig_df
```

```python
plt.plot(trig_df.index,trig_df["sin(x)"], color="purple", label="Sine")
plt.plot(trig_df.index,trig_df["cos(x)"], color="red", label="Cosine", marker="o", linewidth=0)
# plt.plot(trig_df.index,trig_df["tan(x)"])

plt.axis([-5.0,4.9,-1,1])

plt.hlines(0,-5, 10, color="black", alpha=.75)
plt.vlines(0,-1, 10, color="black", alpha=.75)

plt.legend(loc="upper left")
plt.title("Trigonometric Functions")
plt.xlabel("X Values")
plt.ylabel("Y Values"

plt.grid()
plt.savefig("./trig_chart")
plt.show()
```

```python
departments = ["Sales","Marketing", "Development","Research","Management"]
budgets = [1000,1500, 900, 500, 1100]
max_budget = 1000
```

```python
colors = ["red" if amount > max_budget else "black" for amount in budgets]

plt.bar(departments, budgets, color = colors, edgecolor = "blue")

plt.title("Departments and Budgets")
plt.xlabel("Department")
plt.ylabel("Budget")

plt.axhline(y = max_budget, color="black")

plt.show()
```

```python
explode = [.1 if amount > max_budget else 0 for amount in budgets]
plt.pie(budgets, labels=departments, autopct="%1.1f%%", explode = explode, shadow = True, startangle=50)
plt.show()
```

```python
import requests
```

```python
lat = 40.706
lon = -74.5493

loc = requests.get(f"https://api.weather.gov/points/{lat},{lon}").json()
forecast = loc["properties"]["forecast"]
```

```python
data = requests.get(forecast).json()
temps = []
times = []
for day in data["properties"]["periods"]:
    temps.append(day["temperature"])
    times.append(day["name"])
```

```python
plt.figure(figsize=(12,12), dpi = 80)
plt.plot(times, temps, label="Fahrenheit")
plt.plot(times, [(5/9) * (temp - 32) for temp in temps], label="Celsius")
plt.legend()
plt.show()
```
